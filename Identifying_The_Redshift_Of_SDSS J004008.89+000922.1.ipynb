```python
#Importing necessary packages 
import numpy as np 
import matplotlib.pyplot as plt
import scipy.signal
from astropy.io import fits
from Functions import *
from scipy.optimize import curve_fit 
```

```python
hdul = fits.open("spec-0392-51793-0587.fits")
print(hdul[1].columns)
```

```python
spec =  hdul[1].data['flux']
waves = 10**hdul[1].data['loglam']
plt.plot(waves, spec)
```

```python
fig = plt.figure()
fig.set_size_inches(25,10)
plt.plot(waves, spec)
plt.show()
```

```python
sub_i = np.where((waves > 5000) & (waves < 5500))
sub_spec = spec[sub_i]
sub_wave = waves[sub_i]
fig = plt.figure()
fig.set_size_inches(25, 10)
plt.plot(sub_wave, sub_spec)
plt.show()
```

```python
sub_i = np.where((waves > 5240) & (waves < 5260))
sub_spec = spec[sub_i]
sub_wave = waves[sub_i]
fig = plt.figure()
fig.set_size_inches(25, 10)
plt.plot(sub_wave, sub_spec)
plt.show()
```
